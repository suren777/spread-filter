# Simulate Futures prices

Considering 2 scenarios:

1. Rolling Futures
2. Not Rolling Futures

```python
x0, alpha, lmbd, sigma, kappa, dt = 4, 0.09, 0.37, 0.43, 0.03, 1/252.
c1,c2,c3,c4 = 1.2,-0.19,0.1,-0.34
eta = 0
n_steps = 252
Tn=[0.3, 0.6]
```

```python
price_path = generate_futures_path(x0, alpha, lmbd, sigma, kappa, dt, Tn, c1, c2, c3, c4, eta, n_steps)
plt.plot(price_path[0].T,label='Spot')
plt.plot(price_path[1].T)
plt.legend()
```

```python
price_path = generate_futures_path(x0, alpha, lmbd, sigma, kappa, dt, Tn, c1, c2, c3, c4, eta, n_steps,rolling=False)
plt.plot(price_path[0].T,label='Spot')
plt.plot(price_path[1].T,label='Futures')
plt.legend()
```
